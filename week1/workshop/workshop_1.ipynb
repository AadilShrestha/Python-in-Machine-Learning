```python
import numpy as np
import random
from timeit import timeit
def add_python(z1,z2):
  for z1,z2 in zip(z1,z2):
    return z1+z2;
def add_numpy(z1,z2):
  return np.add(z1,z2);
size= 100000;
z1 = random.sample(range(size + 1),100)
z2 = random.sample(range(size + 1),100)
python_time = timeit("add_python(z1,z2)",globals=globals(),number=10000)
numpy_time = timeit("add_numpy(z1,z2)",globals=globals(),number=10000)
print("python time",python_time)
print("numpy time",numpy_time)
```

```python
def factorial(n):
    if n==0:
        return 1
    else:
        return n* factorial(n-1)
print(factorial(5))
```

```python
 # 2.1
def sum_nested_list(nested_list):
    """
    Calculate the sum of all numbers in a nested list.
    This function takes a list that may contain integers and other nested lists.
    It recursively traverses the list and sums all the integers, no matter how deeply
    nested they are.
    Args:
    nested_list (list): A list that may contain integers or other lists of integers.
    Returns:
    """
    total = 0
    for element in nested_list:
        if isinstance(element, list): 
            total += sum_nested_list(element) 
        else:
            total += element 
    return total
print("Sum of nested list: ", sum_nested_list([1, [2, [3, 4], 5], 6, [7, 8]]))
```

```python
def generate_permutations(s):
    """
    Generate all unique permutations of a string using recursion.
    
    Args:
        s (str): The input string.
    
    Returns:
        list: A list of unique permutations of the string.
    """
    if len(s) == 1:  # Base case: Single character
        return [s]
    
    permutations = []
    for i, char in enumerate(s):
        remaining = s[:i] + s[i+1:] 
        for perm in generate_permutations(remaining):  
            new_perm = char + perm
            if new_perm not in permutations:  
                permutations.append(new_perm)
    
    return permutations

print("Permutation: ",generate_permutations("abc"))
```

```python
def calculate_directory_size(directory):
    """
    Calculate the total size of a directory, including nested subdirectories.
    Args:
        directory (dict): A dictionary where keys are file names (with sizes in KB)
                          or subdirectory names (with values as another dictionary).
    Returns:
        int: Total size of the directory in KB.
    """
    total = 0
    for key, value in directory.items():
        if isinstance(value, dict):  
            total += calculate_directory_size(value)
        else:  # It's a file
            total += value
    return total
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
    "subsubdir1": {
        "file5.txt": 250
    },
    "file6.txt": 150
    }
}
print("Total size: ", calculate_directory_size(directory_structure))
```

```python
def min_coins(coins, amount):
    """
    Finds the minimum number of coins needed to make up a given amount using dynamic programming.
    
    Args:
        coins (list of int): A list of coin denominations available for making change.
        amount (int): The target amount to make up using the minimum number of coins.
    
    Returns:
        int: The minimum number of coins required to make the given amount. 
             Returns -1 if it is not possible to make the amount with the given coins.
    
    Example:
    >>> min_coins([1, 2, 5], 11)
    3
    >>> min_coins([2], 3)  
    -1
    """
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0 
    

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    

    return dp[amount] if dp[amount] != float('inf') else -1


print(min_coins([1, 2, 5], 11))
print(min_coins([2], 3))        
print(min_coins([1, 3, 4], 6))  
```

```python
def longest_common_subsequence(s1, s2):
    """
    Finds the length of the longest common subsequence (LCS) between two strings using DP.
    
    Args:
        s1 (str): The first string.
        s2 (str): The second string.
    
    Returns:
        int: The length of the LCS.
    
    Example:
    >>> longest_common_subsequence("abcde", "ace")
    3
    >>> longest_common_subsequence("abc", "def")
    0
    """
    n, m = len(s1), len(s2)
    dp = [[0] * (m + 1) for _ in range(n + 1)]
    
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else: 
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    
    return dp[n][m]

print(longest_common_subsequence("abcde", "ace"))  # Output: 3 ("ace")
print(longest_common_subsequence("abc", "def"))    # Output: 0
print(longest_common_subsequence("abc", "abc"))    # Output: 3 ("abc")
```

```python
def knapsack(weights, values, capacity):
    """
    Solves the 0/1 Knapsack problem using dynamic programming (tabulation).
    
    Args:
        weights (list of int): The weights of the items.
        values (list of int): The values of the items.
        capacity (int): The maximum weight capacity of the knapsack.
    
    Returns:
        int: The maximum value that can be achieved within the weight capacity.
    
    Example:
    >>> knapsack([1, 3, 4, 5], [1, 4, 5, 7], 7)
    9
    """
    n = len(weights)
    
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:  
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else: 
                dp[i][w] = dp[i - 1][w]
    
    return dp[n][capacity]

weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7

print(knapsack(weights, values, capacity))
```
